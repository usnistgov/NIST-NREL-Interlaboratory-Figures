# Imports and Functions

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np
```

# Get the data

```python
NIST_data = pd.read_csv('xrd-peaks-NIST.csv')
NREL_data = pd.read_csv('xrd-peaks-NREL.csv')
```

```python
colors = 'krgb'
font = {'size'   : 18}
matplotlib.rc('font', **font)
plt.figure(figsize=(10,10))

plt.scatter(NIST_data['2015.1 (NREL)'],NIST_data['2015.1 (NIST)'],c=colors[0], label = 'lib.2015.1')
plt.scatter(NREL_data['2016.1 (NREL)'],NREL_data['2016.1 (NIST)'],c=colors[1], label = 'lib.2016.1')
plt.scatter(NREL_data['2016.2 (NREL)'],NREL_data['2016.2 (NIST)'],c=colors[2], label = 'lib.2016.2')
plt.scatter(NREL_data['2016.3 (NREL)'],NREL_data['2016.3 (NIST)'],c=colors[3], label = 'lib.2016.3')

plt.plot([0,50],[0,50],'k--')
plt.legend()
plt.axes().set_aspect('equal')
ax = plt.axes()

plt.xlim(32, 39)
plt.ylim(32, 39)
plt.grid(True)
plt.xlabel('NREL FSDP (2-theta)')
plt.ylabel('NIST FSDP (2-theta)')
plt.savefig(".pdf", bbox_inches='tight')
plt.show()
```
